actualizar productos en el inventario, así como realizar consultas y generar informes sobre el

```python
class Inventory:

    #Constructor
    def __init__(self):
        self._product_list = {}


    #metodo para agregar producto
    def add_product(self, product):
        product_id = product.get_product_id()
        print("id del producto: ", product_id)
        print("producto: ", product.get_product_name())
        self._product_list.update({ product_id : product })

    #metodo para eliminar producto
    def delete_product(self, product_id):
        self._product_list.pop(product_id)

    #metodo para buscar producto por id
    def search_product_by_id(self, product_id):
        return self._product_list[product_id]

    #metodo para actualizar el producto
    def update_product_id(self, prev_value, product):
        self._product_list.pop(prev_value)
        self._product_list[product.get_product_id()] = product

    #metodo para listar todo, v1 solo devuelve la lista
    def find_all(self):
        return self._product_list
```

```python
class Product:

    #variable global para id unicos y autoincrementables
    _global_id = 0

    #Constructor:
    def __init__(self, product_name, product_quantity, product_price):
        self._product_id = Product._global_id
        self._product_name = product_name
        self._product_quantity = product_quantity
        self._product_price = product_price
        Product._global_id += 1

    #access methods for id
    def get_product_id(self):
        return self._product_id

    def set_product_id(self, value):
        self._product_id = value

    #access methods for name
    def get_product_name(self):
        return self._product_name

    def set_product_name(self, value):
        self._product_name = value

    #access methods for quantity
    def get_product_quantity(self):
        return self._product_quantity

    def set_product_quantity(self, value):
        self._product_quantity = value

    #access methods for price
    def get_product_price(self):
        return self._product_price

    def set_product_price(self, value):
        self._product_price= value
```

```python
inventory = Inventory()
first_time_flag = True
total_price = 0
total_quantity = 0
```

```python
def menu_handler(option):

    items = inventory.find_all()
    item_ids = []

    def show_products():
        for item in items.values():
            global first_time_flag
            global total_price
            global total_quantity
            product_id = item.get_product_id()
            if first_time_flag or len(item_ids) == 0:
                item_ids.append(product_id)
                first_time_flag = False
            item_ids.append(product_id)
            product_name = item.get_product_name()
            product_quantity = item.get_product_quantity()
            total_quantity += product_quantity
            product_price = item.get_product_price()
            total_price += product_price

            product_text = "id: {} / nombre: {} / cantidad: {} / precio: {}".format(product_id, product_name, product_quantity, product_price)
            print(product_text)


    if option == 1:

        valid_quantity = False
        valid_price = False
        input_error = "datos incorrectos, intente nuevamente"

        print("ingrese el nombre del producto:")
        product_name = input()

        while not valid_quantity and not valid_price:

            while not valid_quantity:

                print("ingrese la cantidad de producto:")

                try:
                    product_quantity = int(input())
                except:
                    print(input_error)
                else:
                    if product_quantity < 0:
                        print(input_error)
                    else:
                        valid_quantity = not valid_quantity

            while not valid_price:

                print("Ingrese el precio del producto")

                try:
                    product_price = int(input())
                except:
                    print(input_error)
                else:
                    if product_price < 0:
                        print(input_error)
                    else:
                        valid_price = not valid_price

            product = Product(product_name, product_quantity, product_price)
            inventory.add_product(product)
    elif option == 2:
        if len(items) == 0:
            print("No existen productos en el inventario, intentelo mas tarde")
        else:
            show_products()
            print("Ingrese id del producto a eliminar:")
            valid_product_id = False
            while not valid_product_id:
                try:
                    input_product_id = int(input())
                except:
                    print("Datos erroneos, intente nuevamente")
                else:
                    if input_product_id not in item_ids:
                        print("El id ingresado no existe, intente nuevamente")
                    else:
                        inventory.delete_product(input_product_id)
                        print("Producto eliminado")
                        valid_product_id = not valid_product_id
    elif option == 3:
        if len(items) == 0:
            print("No existen productos en el inventario, intentelo mas tarde")
        else:
            valid_id = False
            show_products()
            while not valid_id :
                try:
                    print("Ingrese id del producto a actualzar:")
                    input_product_id = int(input())
                except:
                    print("Datos erroneos, intente nuevamente")
                else:
                    if input_product_id not in item_ids:
                        print("El id ingresado no existe, intente nuevamente")
                    else:
                        valid_id = True
                        valid_product_option = False
                        product = items[input_product_id]
                        print("1 - Id: ", product.get_product_id())
                        print("2 - Nombre: ", product.get_product_name())
                        print("3 - Cantidad: ", product.get_product_quantity())
                        print("4 - Precio: ", product.get_product_price())
                        print("Seleccione opcion:")
                        while not valid_product_option:
                            try:
                                product_option = int(input())
                            except:
                                print("Opcion no valida, intente nuevamente")
                            else:
                                if product_option == 1:
                                    prev_product_id = product.get_product_id()
                                    valid_id = False
                                    while not valid_id:
                                        try:
                                            print("ingrese el nuevo id del producto:")
                                            new_product_id = int(input())
                                        except:
                                            print("Datos erroneos, intente nuevamente")
                                        else:
                                            if new_product_id < 0:
                                                print("Seleccione un numero a partir de cero")
                                            else:
                                                product.set_product_id(new_product_id)
                                                inventory.update_product_id(prev_product_id, product)
                                                valid_product_option = not valid_product_option
                                                valid_id = True
                                elif product_option == 2:
                                    print("ingrese el nuevo nombre del producto:")
                                    new_product_name = input()
                                    product.set_product_name(new_product_name)
                                    valid_product_option = not valid_product_option
                                elif product_option == 3:
                                    valid_product_quantity = False
                                    while not valid_product_quantity:
                                        try:
                                            print("ingrese la nueva cantidad del producto:")
                                            new_product_quantity = int(input())
                                        except:
                                            print("Datos incorrectos, intentelo nuevamente")
                                        else:
                                            if new_product_quantity < 0:
                                                print("Seleccione un numero a partir de cero")
                                            else:
                                                product.set_product_quantity(new_product_quantity)
                                                valid_product_quantity = not valid_product_quantity
                                                valid_product_option = not valid_product_option
                                else:
                                    valid_product_price = False
                                    while not valid_product_price:
                                        try:
                                            print("ingrese el nuevo precio del producto:")
                                            new_product_price = int(input())
                                        except:
                                            print("Datos incorrectos, intentelo nuevamente")
                                        else:
                                            product.set_product_price(new_product_price)
                                            valid_product_price = not valid_product_price
                                            valid_product_option = not valid_product_option
    elif option == 4:
        valid_id = False
        id_found = False
        while not valid_id:
            try:
                print("Ingrese el id del producto a buscar:")
                search_id = int(input())
            except:
                print("Datos incorrectos, ingrese nuevamente")
            else:
                if search_id not in items.keys():
                    print("Id no encontrado, intente nuevamente")
                else:
                    id_found = not id_found
                    valid_id = not valid_id
                    product = items[search_id]
                    print("Id: ", product.get_product_id())
                    print("Nombre: ", product.get_product_name())
                    print("Cantidad:", product.get_product_quantity())
                    print("Precio: ", product.get_product_price())

    else:
        print("Lista de productos: ")
        show_products()
        print("******************************")
        print("Cantidad total de productos: ", total_quantity)
        print("******************************")
        print("Precio total acumulado: ", total_price)



```

```python
def menu():

    menu_flag = False


    def show_menu():
        print("******************************")
        print("Bienvenido al sistema de inventario 31M")
        print("Elija una opcion:")
        print("1 - Agregar producto")
        print("2 - Eliminar producto")
        print("3 - Actualizar producto")
        print("4 - Encontrar producto")
        print("5 - Listar todo")
        print("6 - Salir")
        print("******************************")

    error_message = "Opcion no valida, intente nuevamente"

    while not menu_flag:

        show_menu()

        try:
            option = int(input())
        except ValueError:
            print(error_message)
        else:
            if option in range(1, 7):
                menu_handler(option)
                if option == 6:
                    print("Hasta pronto!")
                    menu_flag = not menu_flag
            else:
                print(error_message)

menu()
```
